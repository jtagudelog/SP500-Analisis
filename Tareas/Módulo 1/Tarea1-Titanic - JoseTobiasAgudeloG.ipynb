```python
#5.Mostrar por pantalla el porcentaje de personas que sobrevivieron y murieron.
c = df['Survived'].count()
c2 = df['Survived'].value_counts()

for i in c2.items():
    p = (i[1] / c) * 100
    o = i[0]
    print("El porcentaje de {} es {:.2f}%".format(o, p))
```

```python
#8. Calcula el porcentaje de valores nulos en cada columna.
d=df.columns
sum={}
for i in d:
  if i not in sum:
    sum[i]=0
  h= df[i].isnull().sum()
  sum[i] += h
total={}
for i in d:
     total[i]=df[i].count()
por={}
for i in d:
  por[i]=(sum[i]/total[i])*100

print(por)
```

```python
#9. Agrupa los datos por el puerto de embarque ('Embarked') y calcula el
#número total de pasajeros y la tarifa promedio ('Fare') para cada grupo.
df.groupby(['Embarked']).agg({"PassengerId":np.count_nonzero,"Fare":np.mean})
```

```python
#10.Calcula la desviación estándar de la tarifa ('Fare') dentro de cada grupo de
#puerto de embarque ('Embarked').
df.groupby(['Embarked']).agg({"Fare":np.std})
```
